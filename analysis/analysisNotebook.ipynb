```python
import csv
import pandas as pd
import numpy as np
import datetime
```

```python
basePath = "../benchmark-client/app/"
```

```python
import glob
msgs = glob.glob(basePath + "/messages/*")
results = glob.glob(basePath + "/results/*")
```

```python
## For Messages
msgsList = []

for filename in msgs:
    df = pd.read_csv(filename, index_col=None, header=0, delimiter=";")
    msgsList.append(df)

msgDf = pd.concat(msgsList, axis=0, ignore_index=True)
```

```python
## For Resulsts
resultsList = []

for filename in results:
    df = pd.read_csv(filename, index_col=None, header=0, delimiter=";")
    resultsList.append(df)

resultDf = pd.concat(resultsList, axis=0, ignore_index=True)
```

```python
unqiueResults = resultDf.drop_duplicates(subset='windowStart', keep='last')
```

```python
unqiueResults
```

```python
unqiueResults.reset_index(drop=True, inplace=True)

for index, row in unqiueResults.iterrows():
    arrayStr = row['composed'].replace("[", "").replace("]", "")
    data = arrayStr.split(",")

    if data[0] == '0':
        continue
        
    print(row['writeOutTimestamp'], row['composed'])
    
    writeOut = datetime.datetime.fromtimestamp(row['writeOutTimestamp']/1000.0)
        
    x = []
    for el in data:
        sendData = msgDf.loc[msgDf['id'] == el]["sendTimestamp"].values[0]
        x.append(sendData)
        
    writeInMax = datetime.datetime.fromtimestamp(np.max(x)/1000.0)
    writeInMin = datetime.datetime.fromtimestamp(np.min(x)/1000.0)
    
    print(writeOut - writeInMax - datetime.timedelta(seconds=5))
    print(writeOut - writeInMin - datetime.timedelta(seconds=5))
    
```
